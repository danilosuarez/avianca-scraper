```python
import pandas as pd

df_parquet = pd.read_parquet("avianca_busquedas_local.parquet")

df_filtrado = df_parquet[df_parquet["num_vuelo"] != "001"]

vuelos_por_fecha = (
    df_filtrado
    .groupby("fecha_busqueda")
    .size()
    .reset_index(name="vuelos")
    .sort_values("fecha_busqueda")
)

print(vuelos_por_fecha)
display(df_filtrado)

```
